```python
problem=[]
index=[]
for i in range(len(content)):
    if len(content[i])!=22:
        problem.append([i,len(content[i])])
        index.append(i)
```

```python
more_index=[]
less_index=[]
for a in problem:
    if a[1]>22 and a[1] not in [42,60,61,64,67,80,82,88,89,92,93,97,98] and a[0]<110:
        more_index.append(a[0])
    else:
        less_index.append(a[0])
```
