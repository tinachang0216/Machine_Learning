## Part C Analysis

```python
%%sql
select sum(value_in_000) / (select sum(value_in_000) from trade where partneriso3 = 'WLD' and tradeflow = 'Import' and year='2016') as perc_wld
from trade
where partneriso3 = 'CHN' and tradeflow = 'Import'and year ='2016'
;
```

```python
%%sql
select year, sum(value_in_000) / (select sum(value_in_000) from trade where partneriso3 = 'WLD' and tradeflow = 'Import' GROUP BY year) as perc_wld
from trade
where partneriso3 = 'CHN' and tradeflow = 'Import'
GROUP BY year
ORDER BY year
;
```

```python
%%sql
select year, partneriso3, sum(value_in_000) / (select sum(value_in_000) from trade where partneriso3 = 'WLD' and tradeflow = 'Import' GROUP BY year) as perc_wld
from trade
where partneriso3 = 'CHN' and tradeflow = 'Import'
GROUP BY year
ORDER BY year
;
```

```python
%%sql
WITH CHN AS(
    SELECT year, partneriso3, SUM(value_in_000) AS sum1
    FROM trade
    WHERE partneriso3 = 'CHN' AND tradeflow='Import'
    GROUP by year, partneriso3), 
WLD AS(
    SELECT year, partneriso3, SUM(value_in_000) AS sum2
    FROM trade
    WHERE partneriso3 = 'WLD' AND tradeflow='Import'
    GROUP by year, partneriso3)
SELECT wld.year, 100*sum1/sum2 AS perc FROM CHN, WLD WHERE CHN.year=WLD.year order by year
```

```python
%%sql
WITH lowincome AS(
    SELECT product.productgroupdes, country.incomegroup, trade.year, SUM(trade.value_in_000) AS sum1
    FROM product, country, trade
    WHERE product.productgroupdes = 'Raw materials' AND incomegroup='High income' AND tradeflow='Import' 
    AND trade.productcode = product.productcode
    GROUP by product.productgroupdes, country.incomegroup, trade.year), 
WLD AS(
    SELECT product.productgroupdes, country.incomegroup, trade.year, SUM(trade.value_in_000) AS sum2
    FROM product, country, trade
    WHERE product.productgroupdes = 'Raw materials' AND tradeflow='Import' AND trade.productcode = product.productcode
    GROUP by product.productgroupdes, country.incomegroup, trade.partneriso3, trade.year)
SELECT WLD.year, lowincome.productgroupdes, lowincome.incomegroup, 100*lowincome.sum1/WLD.sum2 AS perc FROM lowincome, WLD WHERE lowincome.year=WLD.year order by year  
```

```python
%%sql
WITH lowincome AS(
    SELECT product.productgroupdes, country.incomegroup, trade.year, SUM(trade.value_in_000) AS sum1
    FROM product, country, trade
    WHERE product.productgroupdes = 'Raw materials' AND incomegroup='High income' AND tradeflow='Import' 
    AND trade.productcode = product.productcode
    GROUP by product.productgroupdes, country.incomegroup, trade.year), 
WLD AS(
    SELECT product.productgroupdes, country.incomegroup, trade.year, SUM(trade.value_in_000) AS sum2
    FROM product, country, trade
    WHERE product.productgroupdes = 'Raw materials' AND tradeflow='Import' AND trade.productcode = product.productcode
    GROUP by product.productgroupdes, country.incomegroup, trade.partneriso3, trade.year)
SELECT WLD.year, lowincome.productgroupdes, lowincome.incomegroup, 100*lowincome.sum1/WLD.sum2 AS perc FROM lowincome, WLD WHERE lowincome.year=WLD.year order by year  
```
