```python
from lxml import html
import requests
import pandas as pd
```

```python
name=[]
for i in range(1,610):
    j = str(i)
    page = requests.get('http://www.brazil4export.com/en/pesquisa/resultado/?page='+j+'&countries=249')
    tree = html.fromstring(page.content)
    name.extend(tree.xpath('//div[@class="panel-heading"]/@data-name'))
```

```python
address =[]
for i in range(1,610):
    j = str(i)
    page = requests.get('http://www.brazil4export.com/en/pesquisa/resultado/?page='+j+'&countries=249')
    tree = html.fromstring(page.content)
    address.extend(tree.xpath('//div[@class="panel-heading"]/@data-address'))

phones =[]
for i in range(1,610):
    j = str(i)
    page = requests.get('http://www.brazil4export.com/en/pesquisa/resultado/?page='+j+'&countries=249')
    tree = html.fromstring(page.content)
    phones.extend(tree.xpath('//div[@class="panel-heading"]/@data-phones'))

emails =[]
for i in range(1,610):
    j = str(i)
    page = requests.get('http://www.brazil4export.com/en/pesquisa/resultado/?page='+j+'&countries=249')
    tree = html.fromstring(page.content)
    emails.extend(tree.xpath('//div[@class="panel-heading"]/@data-emails'))

site =[]
for i in range(1,610):
    j = str(i)
    page = requests.get('http://www.brazil4export.com/en/pesquisa/resultado/?page='+j+'&countries=249')
    tree = html.fromstring(page.content)
    site.extend(tree.xpath('//div[@class="panel-heading"]/@data-site'))

contact =[]
for i in range(1,610):
    j = str(i)
    page = requests.get('http://www.brazil4export.com/en/pesquisa/resultado/?page='+j+'&countries=249')
    tree = html.fromstring(page.content)
    contact.extend(tree.xpath('//div[@class="panel-heading"]/@data-contact'))

rangeinfo =[]
for i in range(1,610):
    j = str(i)
    page = requests.get('http://www.brazil4export.com/en/pesquisa/resultado/?page='+j+'&countries=249')
    tree = html.fromstring(page.content)
    rangeinfo.extend(tree.xpath('//div[@class="panel-heading"]/@data-range'))

activity =[]
for i in range(1,610):
    j = str(i)
    page = requests.get('http://www.brazil4export.com/en/pesquisa/resultado/?page='+j+'&countries=249')
    tree = html.fromstring(page.content)
    activity.extend(tree.xpath('//div[@class="panel-heading"]/@data-activity'))
```

```python
data_frame = pd.DataFrame(name,columns=['name'])
data_frame['address'] = pd.Series(address, index=data_frame.index)
data_frame['phones'] = pd.Series(phones, index=data_frame.index)
data_frame['emails'] = pd.Series(emails, index=data_frame.index)
data_frame['site'] = pd.Series(site, index=data_frame.index)
data_frame['contact'] = pd.Series(contact, index=data_frame.index)
data_frame['rangeinfo'] = pd.Series(rangeinfo, index=data_frame.index)
data_frame['activity'] = pd.Series(activity, index=data_frame.index)
```
